```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from decimal import *
```

```python
getcontext().prec = 30
def logistic(L,x):
    return 4*L*x*(1-x)
```

```python
n = 1000
L = np.linspace(0.4,1.0,n)
```

```python
iterations = 100000
keep = 10000
```

```python
x = 1e-4 * np.ones(n)
lyapunov = np.zeros(n)
```

```python
plt.figure(figsize=(10,6))
for i in range(iterations):
    lyapunov += np.log(abs(4*L*(1-2*x)))
    x = logistic(L,x)
    
    if i >= (iterations - keep):
        plt.plot(L,x,'.k', markersize=0.1)    
```

```python
plt.figure(figsize=(10,6))
plt.plot(L[lyapunov < 0],lyapunov[lyapunov < 0]/ iterations, '.k')
plt.plot(L[lyapunov >= 0], lyapunov[lyapunov >= 0] / iterations, '.r')
```

```python
Bifure = []
for i in range(len(lyapunov)):
    if abs(lyapunov[i]/iterations)<0.00175:
        Bifure.append(L[i])
```

```python
Bifure = np.asanyarray(Bifure)
```

```python
Figen = Decimal(4*(Bifure[1] - Bifure[0])/(Bifure[2] - Bifure[1]))
```

```python
print(Figen)
```
